```python
import pandas as pd 
import numpy as np
```

```python
def get_data(size=10_000_000):
    df=pd.DataFrame()
    df['age']=np.random.randint(18, 100, size=size)
    df['Time_in_bed']=np.random.randint(0, 9, size=size)
    df['favorite food']=np.random.choice(['pizza', 'pasta', 'burger', 'sandwich', 'salad'], size=size)
    df['hate food']=np.random.choice(['broccoli', 'carrot', 'spinach', 'cabbage', 'cauliflower'], size=size)
    df['name']=np.random.choice(['jhon','tom','jane','siri','alexa','dubo'],size=size)
    return df
```

```python
# %%timeit
get_data()
```
